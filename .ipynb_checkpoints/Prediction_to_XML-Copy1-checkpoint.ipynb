```python
import pandas as pd
import os
import shutil
import numpy as np
```

```python
def create_elem(src_id, tgt_id):
    elem = '<map>\n<Cell>\n<entity1 rdf:resource="'+src_id+'"/>\n'
    elem = elem + '<entity2 rdf:resource="'+tgt_id+'"/>\n<relation>=</relation>\n'
    elem = elem + '<measure rdf:datatype="xsd:float">1.0</measure>\n</Cell>\n</map>'
    return elem
```

```python
path = 'C:/Users/D072202/Desktop/result_data/oaei'
for d in [d for r, d, f in os.walk(path) if not d==[]][0]:
    resultdir = 'C:/Users/D072202/oaei/results/'+d
    outputdir = path + '/' + d
    
    if os.path.exists(resultdir):
        shutil.rmtree(resultdir)
    os.mkdir(resultdir)
    
    predictionfile = open(outputdir+'/prediction.csv')
    ids = pd.read_csv(outputdir+'/test_simple-strcombined_ids.csv')
    prediction = predictionfile.read().replace("[","").replace("]","").replace("\n","").split(" ")
    prediction = np.array(prediction)
    predictionfile.close()
    ids = ids.iloc[np.where(prediction=='1')]
    
    namesdict1 = pd.read_csv('C:/Users/D072202/RData2Graph/rdata2graph/data/oaei_data/instanceID-to-verticeID_darkscape.txt', sep='\t', header=None)
    namesdict2 = pd.read_csv('C:/Users/D072202/RData2Graph/rdata2graph/data/oaei_data/instanceID-to-verticeID_oldschoolrunescape.txt', sep='\t', header=None)
    outfile = open(resultdir+'/darkscape~oldschoolrunescape~results.xml', 'w+')
    starttag = '<?xml version="1.0" encoding="utf-8"?>\n<rdf:RDF xmlns="http://knowledgeweb.semanticweb.org/heterogeneity/alignment"\n  xmlns:rdf="http://www.w3.org/1999/02/22-rdf-syntax-ns#"\n  xmlns:xsd="http://www.w3.org/2001/XMLSchema#">\n<Alignment>\n  <xml>yes</xml>\n  <level>0</level>\n  <type>??</type>\n  <onto1>\n    <Ontology rdf:about="darkscape">\n      <location>http://darkscape.wikia.com</location>\n    </Ontology>\n  </onto1>\n  <onto2>\n    <Ontology rdf:about="oldschoolrunescape">\n      <location>http://oldschoolrunescape.wikia.com</location>\n    </Ontology>\n  </onto2>\n'
    endtag = '</Alignment>\n</rdf:RDF>'
    outfile.write(starttag)
    for index, row in ids.iterrows():
        id1 = namesdict1.loc[namesdict1[1]==row['src_id']][0].values[0]
        id2 = namesdict2.loc[namesdict2[1]==row['tgt_id']][0].values[0]
        outfile.write(create_elem(id1, id2)+"\n")
    outfile.write(endtag)
    outfile.close()
    
    namesdict1 = pd.read_csv('C:/Users/D072202/RData2Graph/rdata2graph/data/oaei_data/instanceID-to-verticeID_darkscape.txt', sep="\t", header=None)
    namesdict1.loc[namesdict1[1]=='http://rdata2graph.sap.com/darkscape/#39305552-9983-4f31-8985-52b37152078a'][0].values[0]
    
    for index, row in ids.iterrows():
        print(row['src_id'])
        break
    
    pd.options.display.max_colwidth = 100
    ids
    
```
